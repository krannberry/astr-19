```python
%matplotlib inline 
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 1, 100)

def exponential_function(x):
    return np.exp(x)
```

```python
y = exponential_function(x)

plt.plot(x, y)
plt.xlabel('Time [milliseconds]')
plt.ylabel('Awesomeness')
plt.title('Exponential Function')
```
