```python
import numpy as np
import matplotlib.pyplot as plt
```

```python

n, p = 10, .5
x = np.random.binomial(n, p, 1000)

plt.hist(x, bins=100, alpha=.5, edgecolor='black')

plt.xlabel('Successes')
plt.ylabel('Frequency')
plt.title('Binomial Distribution')

plt.savefig('cj2-10.pdf')

plt.show()
```
